```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


# 2. Tính toán số lượng và tỷ lệ phần trăm
# value_counts(normalize=True) sẽ tính toán tỷ lệ phần trăm (từ 0 đến 1)
risk_counts = df['risk_level'].value_counts(normalize=True) * 100

# Đảm bảo thứ tự hiển thị trên biểu đồ luôn là: Low -> Medium -> High
ordered_categories = ['Low', 'Medium', 'High']
risk_counts = risk_counts.reindex(ordered_categories)

# 3. Cấu hình giao diện biểu đồ (Style chuẩn phong cách báo cáo chuyên nghiệp)
plt.figure(figsize=(9, 6)) # Định dạng kích thước khung hình (Rộng x Cao)
sns.set_theme(style="whitegrid") # Tạo nền lưới trắng sang trọng

# Định nghĩa bảng màu hiện đại (Xanh dương cho Low, Vàng cam cho Medium, Đỏ cho High)
custom_colors = ['#0ea5e9', '#f59e0b', '#ef4444']

# 4. Vẽ biểu đồ cột (Bar plot)
ax = sns.barplot(
    x=risk_counts.index, 
    y=risk_counts.values, 
    palette=custom_colors,
    hue=risk_counts.index, # Tránh cảnh báo từ seaborn phiên bản mới
    legend=False
)

# 5. Hiển thị số liệu phần trăm trực tiếp trên đầu mỗi cột
for p in ax.patches:
    percentage = f'{p.get_height():.2f}%' # Định dạng lấy 2 chữ số thập phân
    x = p.get_x() + p.get_width() / 2 # Lấy tọa độ trung tâm trục X của cột
    y = p.get_height() + 1.5          # Tọa độ trục Y (đẩy lên trên đầu cột một chút)
    ax.annotate(
        percentage, 
        (x, y), 
        ha='center', 
        va='center', 
        fontsize=12, 
        weight='bold', 
        color='#1e293b'
    )

# 6. Thiết lập tiêu đề và nhãn các trục
plt.title('Tỷ Lệ Phần Trăm Các Mức Độ Rủi Ro Sức Khỏe Tâm Thần', fontsize=16, weight='bold', pad=20, color='#0f172a')
plt.xlabel('Mức Độ Rủi Ro (Risk Level)', fontsize=13, weight='bold', labelpad=10)
plt.ylabel('Tỷ Lệ Phần Trăm (%)', fontsize=13, weight='bold', labelpad=10)

# Giới hạn trục Y cao hơn một chút so với cột cao nhất để không bị đè chữ
plt.ylim(0, 100) 

# Làm sạch giao diện: Bỏ đường viền khung phía trên và bên phải
sns.despine(left=True, bottom=True)

# 7. Hiển thị biểu đồ lên màn hình
plt.tight_layout() # Tự động căn chỉnh không bị mất chữ biên
plt.show()
```
